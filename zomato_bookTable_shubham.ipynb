```python
# code for table booking

option = webdriver.ChromeOptions()
option.add_argument("— incognito")
browser = webdriver.Chrome(executable_path=r'F:\scrapping program\chromedriver_win32\chromedriver.exe')

all_restro_name = []
all_address = []
all_cuisine = []
for i in city_name:
    path = 'https://www.zomato.com/'+i+'/restaurants?table_booking=1'
    browser.get(path)
    time.sleep(3)
    if True:
        try:
            total_page = browser.find_elements_by_xpath('//*[@id="search-results-container"]/div[2]/div[1]/div[1]/div/b[2]')[0]
        except IndexError:
            continue
        total_page = int(total_page.text)

        all_restro_name = []
        all_address = []
        all_cuisine = []

        for q in range(1,total_page+1):
        #for q in range(1,2):
            inner_page = 'https://www.zomato.com/'+i+'/restaurants?table_booking=1&page=' + str(q)
            browser.get(inner_page)
            time.sleep(3)

            restro_name = browser.find_elements_by_xpath("//a[@class='result-title hover_feedback zred bold ln24   fontsize0 ']")
            address = browser.find_elements_by_xpath("//div[@class='col-m-16 search-result-address grey-text nowrap ln22']")
            cuisines = browser.find_elements_by_xpath("//span[@class='col-s-11 col-m-12 nowrap  pl0']")

            for n in restro_name:
                all_restro_name.append(n.text)
            for a in address:
                all_address.append(a.text)
            for c in cuisines:
                all_cuisine.append(c.text)

            lis1 = []
            lis2 = []
            lis3 = []

            for l,p,r in zip(all_restro_name,all_address,all_cuisine):
                lis1.append([l])
                lis2.append([p])
                lis3.append([r])


            with open('tableBookingZomato.csv','a',encoding='utf-8',newline='') as f:
                head=['City','restro_Name','restro_Address','cuisine']
                writer=csv.DictWriter(f,fieldnames=head)
                writer.writeheader()
                for x,y,z in zip(lis1,lis2,lis3):
                    writer.writerow({'City':i,'restro_Name':x,'restro_Address':y,'cuisine':z})
            all_restro_name.clear()
            all_address.clear()
            all_cuisine.clear()
        else:
            continue
```
