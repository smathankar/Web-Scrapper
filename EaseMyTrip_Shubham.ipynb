```python
import requests,csv,time
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
city_list = [ 'indore, india'
    
        ]
```

```python
# main code

option = webdriver.ChromeOptions()
option.add_argument("â€” incognito")
browser = webdriver.Chrome(executable_path=r'F:\scrapping program\chromedriver_win32\chromedriver.exe')
path = 'https://hotel.easemytrip.com/'


for ci in city_list:
    browser.get(path)
    browser.find_elements_by_xpath("//input[@class='loc-im ui-autocomplete-input']")[0].clear()
    browser.find_elements_by_xpath("//input[@class='loc-im ui-autocomplete-input']")[0].send_keys(ci)
    
    submit_button = browser.find_elements_by_xpath("//*[@class='srch-m']")[0]
    submit_button.click()
    time.sleep(5)
    
    
    main_city_page_url = browser.current_url
    
    
    
    tot_hot = browser.find_element_by_xpath("//*[@id='htlTotal']")
    tot_hotel = int(tot_hot.text)
    time.sleep(5)
    
    
    # code for scrolling down
    last_height = browser.execute_script("return document.body.scrollHeight")
    while True:
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(0.5)
        new_height = browser.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    
    
    

    
    
    time.sleep(5)
    hotel_url_name = []

    for q in range(1,tot_hotel+1):
        try:
            hot_url_nam = browser.find_element_by_xpath("//*[@id='hotelListDiv']/div/div/div["+str(q)+"]/div[1]/div[2]/h2/a")
            hotel_url_name.append(hot_url_nam.text)
        except:
            continue
    
    
    for x in hotel_url_name:
        choose_account = browser.find_element_by_link_text(""+x+"")
        choose_account.click()
        time.sleep(3)
        
        
        
        hotel_name = []
        hotel_add = []
        hotel_ami = []
        hotel_ami2 = []

        hotname = browser.find_element_by_xpath("//*[@id='myDescForm']/div[3]/div[1]/div[1]/span")
        for hn in hotname:
            hotel_name.append(hn.text)
        hotadd = browser.find_element_by_xpath("//*[@id='myDescForm']/div[3]/div[1]/div[2]/span")
        for ha in hotadd:
            hotel_add.append(ha.text)
        hotami = browser.find_element_by_xpath("//*[@id='move']/div[3]/div/div/div[1]")
        for ham in hotami:
            hotel_ami.append(ham.text)

        vm = browser.find_element_by_link_text("//*[@id='move']/div[3]/div/div/div[4]/span[1]")
        vm.click()
        hotami2 = browser.find_element_by_xpath("//*[@id='move']/div[3]/div/div/div[3]")
        for ham2 in hotami2:
            hotel_ami2.append(ham2.text)
            
            
        hotel_Amenit1=[]
        for i in hotel_ami:
            hotel_Amenit1.append(i)
        hotel_Amenit2=[]
        for j in hotel_ami2:
            hotel_Amenit2.append(j)
        tot_aminity =  hotel_Amenit1 + hotel_Amenit2
        total_aminity = [tot_aminity]
        
        
        
        
        
        #code for csv wriitng
        
        with open('ease My Trip.csv','a',encoding='utf-8',newline='') as f:
            head=['City','hotel_Name','hotel_Address','Aminity']
            writer=csv.DictWriter(f,fieldnames=head)
            writer.writeheader()
            for tamt in total_aminity:
                writer.writerow({'City':ci,'hotel_Name':hotel_name,'hotel_Address':hotel_add,'M1':tamt})

        
        
        browser.get(main_city_page_url)
        del main_city_page_url
print(hotname.text)
print(hotadd.text)
print(hotami.text)
```
